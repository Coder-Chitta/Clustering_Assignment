#### Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.

#### Data Description:
#### Murder -- Muder rates in different places of United States
#### Assualt- Assualt rate in different places of United States
#### UrbanPop - urban population in different places of United States
#### Rape - Rape rate in different places of United States


```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crimedata = pd.read_csv("crime_data.csv")
crimedata
```

```python
crime_data1 = crimedata.rename({'Unnamed: 0':'Location'},axis=1)
crime_data1
```

### Heiarchical Clustering

```python
#Normalization
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
#Normalize Dataframe
```

```python
df_norm = norm_func(crime_data1.iloc[:,1:])
df_norm
```

```python
#Create Dendogram
```

```python
fig = plt.figure(figsize=(20,10))
dendogram = sch.dendrogram(sch.linkage(df_norm,method='single'))
```

```python
fig = plt.figure(figsize=(20,10))
dendogram = sch.dendrogram(sch.linkage(df_norm,method='complete'))
```

```python
fig = plt.figure(figsize=(20,10))
dendogram = sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
fig = plt.figure(figsize=(20,10))
dendogram = sch.dendrogram(sch.linkage(df_norm,method='centroid'))
```

```python
#create clusters
```

```python
hc = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage="ward")
```

```python
#save cluster for chart
```

```python
y_hc = hc.fit_predict(df_norm)
clusters = pd.DataFrame(y_hc,columns=['h_clusters'])
clusters
```

```python
crime_data1['h_clusterid'] = clusters
crime_data1
```

### KMeans

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import k_means
import seaborn as sns
```

```python
#Normalise function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_norm = scaler.fit_transform(crime_data1.iloc[:,1:])
df_norm
```

```python
#Elbow Chart
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime_data1['k_Cluster']=x
crime_data1
```

```python
crime_data1.iloc[:,1:5].groupby(crime_data1.k_Cluster).mean()
```

```python
plt.style.use(['classic']) 
print(plt.style.available) 
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(df_norm)
y_kmeans = kmeans.predict(df_norm)
```

```python
plt.scatter(df_norm[:,0:1],df_norm[:,2:3],c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

## DBSCAN


```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crimedata.info()
```

```python
df_norm
```

```python
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(df_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
crime_data1['db_clusters']=dbscan.labels_
crime_data1
```

```python
crime_data1.groupby('db_clusters').agg(['mean']).reset_index()
```
